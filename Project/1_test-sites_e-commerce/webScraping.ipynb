```python
import requests
from bs4 import BeautifulSoup
import pandas

url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"

r = requests.get(url)
soup = BeautifulSoup(r.text, "lxml")
soup
```

```python
name = soup.find_all('a', class_="title")
product_name = []
for i in name:
    product_name.append(i.text)

print(len(product_name))
product_name
```

```python
price = soup.find_all('h4', class_="price")
product_price = []
for i in price:
    product_price.append(i.text)
print(len(product_price))
product_price
```

```python
description  = soup.find_all('p', class_="description")
product_description = []
for i in description :
    product_description .append(i.text)
print(len(product_description))
product_description
```

```python
rating  = soup.find_all('div', class_="ratings")
# print(rating)
product_rating = []
for i in rating :
    product_rating.append(len(i.find_all('span')))
print(len(product_rating))
product_rating
```

```python
review  = soup.find_all('p', class_="review-count float-end")
# print(rating)
product_Review = []
for i in review :
    product_Review.append(i.text)
print(len(product_Review))
product_Review
```

```python
dic = {
    "product_name" : product_name,
    "product_price" : product_price,
    "product_description" : product_description,
    "product_rating" : product_rating,
    "product_Review" : product_Review
}
df = pandas.DataFrame(dic)
df
```

```python
csv = df.to_csv('output.csv', index=False)
```
